```python
import random, time

a = [random.randint(0,1000) for r in range(500)]
before = time.time()
print a
print 'time: ', time.time()-before
```

```python
# Supposed to be Faster than the regular random
import numpy.random as nprnd 
now = time.time()
print list(nprnd.randint(1000,size=500))
print 'time: ', time.time()-now
```
